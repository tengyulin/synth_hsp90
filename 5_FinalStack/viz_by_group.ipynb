```python
import mrcfile
import matplotlib.pyplot as plt
import numpy as np

# Load the file
with mrcfile.open('/home/danlin/synth_hsp90/hsp90_50PDs_SNR01.mrcs', 'r') as mrc:
    data = mrc.data
```

```python
import random

selected_indices = random.sample(range(data.shape[0]),9)
images = data[selected_indices]
images.shape

fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    # Plot the i-th image
    img = images[i]
    ax.imshow(img, cmap='gray')

    # Remove axis
    ax.axis('off')

plt.show()
```
